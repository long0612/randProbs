```python
nodes = set()
node0 = mst.Node((0,0))
node1 = mst.Node((1,1))
node2 = mst.Node((1,-1))
node3 = mst.Node((2,1))
node4 = mst.Node((2,0))
node5 = mst.Node((2,-1))
node6 = mst.Node((3,1))
node7 = mst.Node((3,-1))
node8 = mst.Node((4,0))
nodes.add(node0)
nodes.add(node1)
nodes.add(node2)
nodes.add(node3)
nodes.add(node4)
nodes.add(node5)
nodes.add(node6)
nodes.add(node7)
nodes.add(node8)

locMap = {}
for node in nodes:
    locMap[node] = node.val

edges = set()
edges.add(mst.Edge(node0,node1,4))
edges.add(mst.Edge(node0,node2,8))
edges.add(mst.Edge(node1,node2,11))
edges.add(mst.Edge(node1,node3,8))
edges.add(mst.Edge(node2,node4,7))
edges.add(mst.Edge(node2,node5,1))
edges.add(mst.Edge(node3,node6,7))
edges.add(mst.Edge(node3,node7,4))
edges.add(mst.Edge(node4,node3,2))
edges.add(mst.Edge(node4,node5,6))
edges.add(mst.Edge(node5,node7,2))
edges.add(mst.Edge(node6,node7,14))
edges.add(mst.Edge(node6,node8,9))
edges.add(mst.Edge(node7,node8,10))
```

```python
mst.visualize(nodes,edges,locMap)
```

```python
nodes = set()
nodes.add(mst.Node((0,0)))
nodes.add(mst.Node((1,0)))
nodes.add(mst.Node((2,-0.2)))
nodes.add(mst.Node((3,0.2)))
nodes.add(mst.Node((0,2)))
nodes.add(mst.Node((2,2)))
nodes.add(mst.Node((3,2)))
nodes.add(mst.Node((5,.8)))
nodes.add(mst.Node((-1,1)))

edges = set()
locMap = {}
for node in nodes:
    locMap[node] = node.val
    for ngb in nodes:
        if ngb != node:
            edgeNew = mst.Edge(node,ngb,np.linalg.norm(np.array(node.val)-np.array(ngb.val)))
            if edgeNew not in edges:
                #print('edgeNew.ePts = %s' % edgeNew.ePts)
                edges.add(edgeNew)
```

```python
mst.visualize(nodes,edges,locMap)
```

```python
edgesP = mst.Prim(nodes,edges)
```

```python
mst.visualize(nodes,edgesP,locMap)
```

```python
edgesK = mst.Kruskal(nodes,edges)
```

```python
mst.visualize(nodes,edgesK,locMap)
```
