```python
import numpy as np
import matplotlib.pyplot as plt
import mst

%load_ext autoreload
%autoreload 2
```

```python
nodes = set()
nodes.add(mst.Node((0,0)))
nodes.add(mst.Node((1,0)))
nodes.add(mst.Node((2,-0.2)))
nodes.add(mst.Node((3,0.2)))
nodes.add(mst.Node((0,2)))
nodes.add(mst.Node((2,2)))
nodes.add(mst.Node((3,2)))
nodes.add(mst.Node((5,.8)))
nodes.add(mst.Node((-1,1)))

edges = set()
locMap = {}
for node in nodes:
    locMap[node] = node.val
    for ngb in nodes:
        if ngb != node:
            edgeNew = mst.Edge(node,ngb,np.linalg.norm(np.array(node.val)-np.array(ngb.val)))
            if edgeNew not in edges:
                #print('edgeNew.ePts = %s' % edgeNew.ePts)
                edges.add(edgeNew)
```

```python
mst.visualize(nodes,edges,locMap)
```

```python
nodesF,edgesF = mst.Prim(nodes,edges)
locMapF = {}
for node in nodesF:
    locMapF[node] = node.val
```

```python
mst.visualize(nodesF,edgesF,locMapF)
```
