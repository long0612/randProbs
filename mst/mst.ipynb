```python
import numpy as np
import matplotlib.pyplot as plt
import mst

%load_ext autoreload
%autoreload 2
```

```python
aGraph = set()
aGraph.add(mst.Node((0,0)))
aGraph.add(mst.Node((1,0)))
aGraph.add(mst.Node((2,0)))
aGraph.add(mst.Node((3,0)))
aGraph.add(mst.Node((0,1)))
aGraph.add(mst.Node((2,1)))
aGraph.add(mst.Node((3,2)))
aGraph.add(mst.Node((4,4)))

locMap = {}
for node in aGraph:
    locMap[node] = node.val
    for ngb in aGraph:
        node.edges[ngb] = np.linalg.norm(np.array(ngb.val)-np.array(node.val))
```

```python
mst.visualize(aGraph,locMap)
```
